### Guys, @radpet asks how many characters are there in the longest sentence

```python
lens = dataset["snippet"].str.len()
lens.mean(), lens.std(), lens.max()
```

```python
%matplotlib inline
lens.hist()
plt.show()
```

### What about the word count?

```python
lens = dataset["snippet"].str.split(" ").str.len()
lens.mean(), lens.std(), lens.max()
```

```python
lens.hist()
plt.show()
```

### What about unique words in train and test provided by our dear friends from Ontotext?

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer().fit(dataset["snippet"].append(onto_test["snippet"]))
len(count_vect.vocabulary_)
```

### You ask yourself what the most common words are?

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
text = ""
for sentence in dataset["snippet"].values:
    text+=sentence
    
wc = WordCloud(background_color="white", max_words=1000,max_font_size= 20,
               stopwords=STOPWORDS).generate_from_text(text)
```

```python
plt.figure(figsize=(15,12))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
